```python
from scipy import stats #import statistics package to run confidence intervals

#general formula for calculating confidence interval of the mean of the data(where data is your chosen values)
#this test will tell me how many values truly represent the computed average (or 95% confidence that the mean is
#between the two numbers given)

def CI_mean(data):
    se = np.std(data)/np.sqrt(len(data))
    low = np.mean(data)-1.96*se
    up = np.mean(data)+1.96*se
    print("Average is:", np.mean(data))
    print("95% CI of mean is: ({}, {})".format(low,up))
```

```python
#imputting my chosen values into the formula
CI_mean(hispanic_value)
```

```python
#conducting significance test to compare whether there is a statistically significant difference 
#between my chosen values and the group average

t_val, p_val = stats.ttest_1samp(hispanic_value, 34.36)

print("Test statistic:", t_val)
print("p-value:", p_val)
```

```python
#Here I will be conducting a 2 way t-test to determine if there is a significant difference between
#values of anxiety and depression in the White and Black populations during covid in the US
```

```python
#finding the mean/CI of each dataset

def CI_mean(data):
    se = np.std(data)/np.sqrt(len(data))
    low = np.mean(data)-1.96*se
    up = np.mean(data)+1.96*se
    print("Average is:", np.mean(data))
    print("95% CI of mean is: ({}, {})".format(low,up))
```

```python
#finding mean and CI of Black population data

CI_mean(black_value)
```

```python
#finding mean and CI of White population data

CI_mean(white_value)
```

```python
#comparison of means with 95% CI

def CI_mean_diff(list1, list2):
    s1 = np.var(list1)
    s2 = np.var(list2)
    n1 = len(list1)
    n2 = len(list2)
    se2 = s1/n1 + s2/n2
    se = np.sqrt(se2)

    diff = np.mean(list1) - np.mean(list2)
    low = diff - 1.96 * se
    up = diff + 1.96 * se
    print("The average difference is:", diff)
    print("The 95% CI is: ({}, {})".format(low, up))
```

```python
#comparing mean differences of Black and White Population with 95% CI

CI_mean_diff(black_value, white_value)
```

```python
#two way t-test to determine if difference in means is statistically significant

t_val, p_val = stats.ttest_ind(black_value,
                               white_value, equal_var=False)

print("Test statistic:", t_val)
print("p-value:", p_val)
```

```python
# These data indicate that there is no significant difference between the Black and White population
# in values of anxiety/depresion during covid in the US (p > 0.05; two sample t-test). The average difference
#between the 2 populations was 3.86 (95% CI [0.43, 7.30]).
```

```python
df3 = pd.read_csv(r'/Users/jankrovatin/Desktop/NHANES.csv')
```

```python
df3
```

```python
#creating a dataframe to view variables of interest
df3 = df3[['Race3', 'Diabetes']]
```

```python
df3.dropna()
```

```python
#creating a frequency table to do chi-squared test

def function(row):
    if row['Diabetes'] == 0.0:
        return 'Healthy'
    elif row['Diabetes'] == 1.0:
        return 'Diabetic'
    else:
        return 'Unknown'
```

```python
df3['Diabetes'] = df3.apply(function, axis=1)
df3
```

```python
#changing the order of my dataframe so that it is in a frequency table

df3 = pd.crosstab(index=df3["Diabetes"], columns=df3["Race3"]) 
df3
```

```python
total = 0
for race in df3.columns:
    total += np.sum(df3[race])
print(total)
```

```python
from scipy import stats

chi2, p, dof, expected = stats.chi2_contingency(df3)

print("chi2:", chi2)
print("p:", p)
print("dof:", dof)
print("expected:", expected)
```

```python
#the chi-squared test suggests that there is an association 
#between rates of diabetes and race chi^2(10, N = 5000)=23.72; p < 0.008
```
